```python
#!/usr/bin/env python
import os
import ConfigParser
import requests
import json
import datetime
import vincent
import logging

logger = logging.getLogger("notebook")
logger.setLevel(logging.DEBUG)
logger.addHandler(logging.StreamHandler())


class GraphiteProxy:

    def __init__(self, config):
        self.graphite_session = config['session']
        self.url = config['graphite_url']

        # Before adding a target add a known working URL for dedicated001 to
        # assets/knockwurst.py
        self.targets = {
            # "load": "",
            # "workflows": "",
            "disk_use": [
                'alias(smartSummarize(sumSeries(net.cloudant.{0[cluster]}.db*.df.srv.used),"{0[granularity]}"), "{0[cluster]} Used disk space (bytes)")',
                'alias(smartSummarize(sumSeries(net.cloudant.{0[cluster]}.db*.df.srv.free),"{0[granularity]}"), "{0[cluster]} Free disk space (bytes)")'
            ],
            "disk_use_v2": [
                'alias(sumSeries(net.cloudant.{0[cluster]}.db*.df.srv.used), "{0[cluster]} Used disk space (bytes)")',
                'alias(sumSeries(net.cloudant.{0[cluster]}.db*.df.srv.free), "{0[cluster]} Free disk space (bytes)")'
            ],
            # "indexes": "",
            # "replications": "",
            "rate": {
                'verb': [],
                'status_code': [],
                # 'end_point': []
            },
            "rate_v2": {
                'verb' : [],
                'status_code': [],
            },
            "wps": 'alias(smartSummarize(sum(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.document_writes,1))),"{0[granularity]}"), "{0[cluster]} Document Writes Per Second")',  # noqa
            "rps": 'alias(smartSummarize(sum(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.database_reads,1))),"{0[granularity]}"), "{0[cluster]} Document Reads Per Second")',  # noqa
            "response_time": 'alias(smartSummarize(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.request_time,1))),"{0[granularity]}"), "{0[cluster]} Response Time (ms)")',  # noqa
            "map_doc": 'alias(smartSummarize(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.couchjs.map_doc,1))),"{0[granularity]}"), "{0[cluster]} Documents per second through map functions")',  # noqa
            "kv_emits": 'alias(smartSummarize(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.couchjs.emits,1))),"{0[granularity]}"), "{0[cluster]} Key:value pairs emitted per second from map functions")',  # noqa
            "wps_v2": 'alias(sum(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.document_writes,1))), "{0[cluster]} Document Writes Per Second")',  # noqa
            "rps_v2": 'alias(sum(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.database_reads,1))), "{0[cluster]} Document Reads Per Second")',  # noqa
            "response_time_v2": 'alias(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.request_time,1))), "{0[cluster]} Response Time (ms)")',  # noqa
            "map_doc_v2": 'alias(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.couchjs.map_doc,1))), "{0[cluster]} Documents per second through map functions")',  # noqa
            "kv_emits_v2": 'alias(averageSeries(nonNegativeDerivative(scaleToSeconds(net.cloudant.{0[cluster]}.db*.couchdb.couchjs.emits,1))), "{0[cluster]} Key:value pairs emitted per second from map functions")',  # noqa

        }
        # In the absence of a DSL...
        self.generate_status_codes_targets()
        self.generate_verb_targets()

    def list_targets(self):
        targets = []
        for k, v in self.targets.items():
            if isinstance(v, dict):
                for kk in v.keys():
                    targets.append('%s/%s' % (k, kk))
            else:
                targets.append(k)
        return targets

    def generate_status_codes_targets(self):
        # using old school %s formatting for this case. leave .format alone.
        target = 'alias(sum(scaledNonNegativeDerivative(keepSomeLastValues(net.cloudant.{0[cluster]}.db*.couchdb.httpd_status_codes.%s,3),"{0[granularity]}")), "{0[cluster]} %s")'  # noqa
        target_v2 = 'alias(sum(nonNegativeDerivative(keepSomeLastValues(net.cloudant.{0[cluster]}.db*.couchdb.httpd_status_codes.%s,3))), "{0[cluster]} %s")'

        status_codes = [
            ('2*', '2xx'),
            ('3*', '3xx'),
            ('4*', '4xx'),
            ('5*', '5xx')
        ]

        for status_code in status_codes:
            self.targets['rate']['status_code'].append(target % status_code)
            self.targets['rate_v2']['status_code'].append(target_v2 % status_code)

    def generate_verb_targets(self):
        # using %s for this formatting here, leaving .format alone for later.
        target = 'alias(sum(scaledNonNegativeDerivative(keepSomeLastValues(net.cloudant.{0[cluster]}.db*.couchdb.httpd_request_methods.%s,3),"{0[granularity]}")), "{0[cluster]} %s")'  # noqa
        target_v2 = 'alias(sum(nonNegativeDerivative(keepSomeLastValues(net.cloudant.{0[cluster]}.db*.couchdb.httpd_request_methods.%s,3))), "{0[cluster]} %s")'  # noqa
        for verb in ['GET', 'POST', 'PUT', 'DELETE', 'COPY', 'HEAD']:
            self.targets['rate']['verb'].append(target % (verb, verb))
            self.targets['rate_v2']['verb'].append(target_v2 % (verb, verb))

    def format_metric(self, cluster, target, granularity='15sec', group=None):
        """
        Format the query to graphite for the cluster.
        """
        formatted_target = []

        if group:
            requested_target = self.targets[group][target]
        else:
            requested_target = self.targets[target]

        if isinstance(requested_target, list):
            for m in requested_target:
                formatted_target.append(
                    m.format({'cluster': cluster, 'granularity': granularity})
                )
        else:
            formatted_target.append(
                requested_target.format({'cluster': cluster, 'granularity': granularity})
            )
        return formatted_target

    def get_data(
        self,
        cluster,
        target,
        timeseries={
            'start': '-5min',
            'end': 'now',
            'granularity': '15sec'
        },
        group=None,
        format="json"
    ):
        """
        Make the request to graphite.

        :param cluster: a string, the name of a cluster
        :param target: a string representing the metric to display
        :param format: a string either raw or json
        """
        prepped = self.build_request(cluster, target, timeseries, group, format)
        r = self.graphite_session.send(prepped)
        r.raise_for_status()
        if format == 'json':
            return json.dumps({
                'start': timeseries['start'],
                'end': timeseries['end'],
                'granularity': timeseries['granularity'],
                'target_responses': r.json()
            })
        else:
            return r.text

    def build_request(self, cluster, target, timeseries, group=None, format="json"):
        """
        Build the request to graphite. Factored out for easier testing.

        :param cluster: a string, the name of a cluster
        :param target: a string representing the metric to display
        :param timeseries: a dict containing the start/end params for from/until
        :param format: a string either raw or json
        """

        granularity = timeseries['granularity']
        options = {
            "from": timeseries['start'],
            "until": timeseries['end'],
            "tz": "UTC",
            "target": self.format_metric(cluster, target, granularity, group),
            "format" : format
        }
        request = requests.Request('GET', self.url, params=options)
        print "request: url={0} params={1}".format(request.url, request.params)
        prepped = self.graphite_session.prepare_request(request)
        return prepped
```

```python
   


graphite_session = requests.Session()
vincent.core.initialize_notebook() 

def read_dot_clou():
    """util to grab user and web_password creds from ~/.clou"""
    clou = os.path.expanduser("~/.clou")
    config = ConfigParser.RawConfigParser()
    config.read(clou)
    return (
        config.get("cloudant", "user"),
        config.get("cloudant", "web_password")
        )

GRAPHITE_URL = 'https://graphite.cloudant.com/render/'

graphite_session.auth = read_dot_clou()

graphite_proxy = GraphiteProxy({
    'session': graphite_session,
    'graphite_url': GRAPHITE_URL
})




def index_data(data):
    """convert rate data to multi index dict"""
    result = {}
    for rate in data['target_responses']:
        if 'index' not in result:
            result['index'] = [ x[1] for x in rate['datapoints']]
        result[rate['target']] = [ x[0] for x in rate['datapoints']]
    # normalise time
    start_time = min(result['index'])
    result['index'] = [ x-start_time for x in result['index']  ]
    return result
```
