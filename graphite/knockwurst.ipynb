```python
#!/usr/bin/env python
import os
import ConfigParser
import requests
import json
import datetime
import vincent


from graphite_lite import GraphiteProxy

graphite_session = requests.Session()
vincent.core.initialize_notebook() 

def read_dot_clou():
    """util to grab user and web_password creds from ~/.clou"""
    clou = os.path.expanduser("~/.clou")
    config = ConfigParser.RawConfigParser()
    config.read(clou)
    return (
        config.get("cloudant", "user"),
        config.get("cloudant", "web_password")
        )

GRAPHITE_URL = 'https://graphite.cloudant.com/render/'

graphite_session.auth = read_dot_clou()

graphite_proxy = GraphiteProxy({
    'session': graphite_session,
    'graphite_url': GRAPHITE_URL
})




def index_data(data):
    """convert rate data to multi index dict"""
    result = {}
    for rate in data['target_responses']:
        if 'index' not in result:
            result['index'] = [ x[1] for x in rate['datapoints']]
        result[rate['target']] = [ x[0] for x in rate['datapoints']]
    # normalise time
    start_time = min(result['index'])
    result['index'] = [ x-start_time for x in result['index']  ]
    return result
    
default_rate_data = json.loads(graphite_proxy.get_data('meritage', 'status_code', group='rate'))


timeseries={
    'start': '-1h',
    'end': 'now',
    'granularity': '1min'
}

last_hour_data = json.loads(graphite_proxy.get_data(
    'meritage', 'status_code', timeseries=timeseries, group='rate')
)
```

```python
default_indexed_data = index_data(default_rate_data)
lasthour_indexed_data = index_data(last_hour_data)

    
```

```python

vis2 = vincent.Bar(default_indexed_data, iter_idx='index')
vis2.legend(title='Responses')
vis2.display()
```

```python
vis3 = vincent.Bar(lasthour_indexed_data, iter_idx='index')
vis3.legend(title='Responses')
vis3.display()
```
