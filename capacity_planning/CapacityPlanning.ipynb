```python
import os
import requests 
import pandas 
import posixpath
import vincent
import ConfigParser

vincent.core.initialize_notebook() 

def read_dot_clou():
    """util to grab user and web_password creds from ~/.clou"""
    clou = os.path.expanduser("~/.clou")
    config = ConfigParser.RawConfigParser()
    config.read(clou)
    return (
        config.get("cloudant", "emissions_test_user"),
        config.get("cloudant", "emissions_test_password")
    )

auth = read_dot_clou()
```

```python

DB_URL = "https://sapi-emissions-test.cloudant.com/cluster_status"

resp = requests.get(
    posixpath.join(DB_URL, "_all_docs"), 
    auth=auth, 
    headers={'Content-Type':'application/json'},
    params={"include_docs": "true"}
)

documents = resp.json()['rows']
document_ids = [x['id'] for x in documents]


```

```python
def unroll_document(doc):
    rows = []
    data = doc['doc']
    timestamp = data['timestamp']
    cluster = data['cluster']
    report = data['report'][cluster]
    for db_node, db_data in report.iteritems():
        
        row_template = {
            "cluster": cluster, 
            "t": timestamp,
            "node": db_node
        }
        
        for data_point in db_data:
            if data_point == 'memory':
                continue
            row_template[data_point] = db_data[data_point]
        
        memory = db_data['memory']
        for mem_key, mem_data in memory.iteritems():
            row_template[mem_key] = mem_data 
        rows.append(row_template)
    return rows
        
    
        
        
data_rows = []
for doc in documents:
    data_rows.extend(unroll_document(doc))
    
data_frame = pandas.DataFrame(data_rows)
    
```

```python
meritage = data_frame[data_frame['cluster']=='meritage']
meritage.sort(['t'])

cols = [
    "ets",
    "Processes",
    "Code",
    "Binary",
    "Processes Used",
    "Atom",
    "Other",
    "Atom Used"
]

group = vincent.StackedArea(meritage[cols])
group.colors(brew='Spectral')
group.width=750

group.display()
```
