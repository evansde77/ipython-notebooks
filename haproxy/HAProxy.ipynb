```python
import os
import csv
import requests
import cStringIO
import vincent
import ConfigParser
vincent.core.initialize_notebook() 

URL = "https://{0}.cloudant.com/_haproxy;csv"

def read_dot_clou():
    """util to grab user and web_password creds from ~/.clou"""
    clou = os.path.expanduser("~/.clou")
    config = ConfigParser.RawConfigParser()
    config.read(clou)
    return (
        config.get("cloudant", "haproxy_user"),
        config.get("cloudant", "haproxy_password")
    )

AUTH = read_dot_clou()

INT_FIELDS=[
    'lastchg', 'weight', 'hrsp_5xx', 'pid', 'comp_byp', 'comp_rsp', 
    'check_duration', 'rate', 'wredis', 'slim', 'rate_max', 'hrsp_1xx', 
    'chkfail', 'comp_out', 'comp_in', 'dreq', 'check_code', 'sid', 'bout', 'dresp', 
    'type', 'hrsp_other', 'bin', 'econ', 'lbtot', 'stot', 'wretr', 'iid', 'hrsp_4xx', 
    'hanafail', 'downtime', 'smax', 'qcur', 'eresp', 'cli_abrt', 'srv_abrt', 'scur', 'bck', 'qmax', 
    'hrsp_2xx', 'act', 'chkdown', 'hrsp_3xx'
]

```

```python

def haproxy_monitoring_data(username, auth):
    """
    hit the haproxy endpoint for the user provided over HTTP
    returns the raw csv string wrapped in a file like object
    """
    resp = requests.get(URL.format(username), auth=AUTH)
    resp.raise_for_status()
    csv_io = cStringIO.StringIO(resp.text)
    return csv_io

def process_csv_data(input_str):
    """
    convert the 
    """
    reader = csv.reader(input_str)
    first_row = None 
    for row in reader:
        if reader.line_num == 1:
            first_row = row
            continue 
        elem = dict(zip(first_row, row)) 
        for k, v in elem.iteritems():
            if k in INT_FIELDS:
                try:
                    elem[k] = int(v)
                except ValueError:
                    elem[k] = None
        yield elem
        

        
        
PLOT_KEYS = ['svname','hrsp_2xx', 'hrsp_3xx', 'hrsp_4xx', 'hrsp_5xx', 'hrsp_other']

def format_data(raw, keys):
    plot_data = {}
    for data in process_csv_data(raw):
        for k,v in data.iteritems():
            if k in keys:
                column = plot_data.setdefault(k, [])
                column.append(v)
                
    return plot_data
        

def haproxy_access_for_user(username):
    raw_data = haproxy_monitoring_data(username, AUTH)
    plot_data = format_data(raw_data, PLOT_KEYS)
    return plot_data


def plot_user(username):
    user_data = haproxy_access_for_user(username)
    viz = vincent.GroupedBar(user_data, iter_idx='svname')
    viz.legend(title='HAProxy Stats For {0}'.format(username))
    viz.display()


plot_user('evansde77')
plot_user('meritage')
plot_user('kocolosk')
```
